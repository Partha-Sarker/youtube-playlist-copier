```python
from googleapiclient.discovery import build
from Google import Create_Service
```

```python
CLIENT_SECRET_FILE  =  'client_secret.json'
API_NAME  =  'youtube'
API_VERSION  =  'v3'
SCOPES  = ['https://www.googleapis.com/auth/youtube']

service =  Create_Service(CLIENT_SECRET_FILE, API_NAME, API_VERSION, SCOPES)
```

```python
TARGET_PLAYLIST_ID = "PLbR5tqSe3pAQgIN0lEE9Iz54Isw7-zTnX"
CURRENT_VIDEO_ID = "S5NLzfQ1YLQ"
```

```python
response = service.videos().list(part="contentDetails", myRating="like", maxResults=50).execute()
likeListItems = response['items']
nextPageToken = response.get('nextPageToken')

while nextPageToken:
    response = service.videos().list(
        part='contentDetails',
        myRating="like",
        maxResults=50,
        pageToken=nextPageToken
    ).execute()
 
    likeListItems.extend(response['items'])
    nextPageToken = response.get('nextPageToken')
    
likeListItems.reverse()
```

```python
test_list = likeListItems.copy()
for i in range(len(test_list)):
    if test_list[i]['id'] == CURRENT_VIDEO_ID:
        break
del test_list[:i+1]
len(likeListItems)
```

```python
count = 0
for video in test_list:
    count += 1
    request_body = {
        'snippet': {
            'playlistId': TARGET_PLAYLIST_ID,
            'resourceId': {
                'kind': video['kind'],
                'videoId': video['id']
            }
        }
    }
 
    service.playlistItems().insert(
        part='snippet,status',
        body=request_body
    ).execute()
    print(count)
```

```python
def create_playlist(playlist_name):
    request = service.playlists().insert(
    part="snippet,status",
    body={
            "snippet": {
                "title": playlist_name,
                "defaultLanguage": "en"
            },
            "status": {
                "privacyStatus": "private"
            }
        }
    )
    return request.execute()
```

```python
def delete_playlist(playlist_id):
    request = service.playlists().delete(
        id=playlist_id
    )
    try:
        request.execute()
    except:
        print("Playlist not found")
```
